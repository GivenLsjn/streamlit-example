```python
import streamlit as st
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.pyplot as plt
from PIL import Image
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import datetime as dt
from dateutil.relativedelta import relativedelta
import altair as alt
```

```python
from google.colab import files
uploaded = files.upload()
data = pd.read_excel("alternative1.xlsx", header=0) # index_col="DATE")
data['DATE']= pd.to_datetime(data['DATE']) #convert "DATE" to a date time
#data.head()
```

```python
st.set_page_config(page_title="G01 Run Out", layout="wide")

cols1,_ = st.columns((1,2)) # To make it narrower
format = 'DD MM, YYYY'  # format output
start_date = dt.date(year=2023,month=11,day=1)-relativedelta(years=2)  #  I need some range in the past
end_date = dt.datetime.now().date()-relativedelta(years=2)
max_days = end_date-start_date

slider = cols1.slider('Select a date range', min_value=start_date, value=end_date ,max_value=end_date, format=format)
## Sanity check
st.table(pd.DataFrame([[start_date, slider, end_date]],
                      columns=['start',
                               'selected',
                               'end'],
                      index=['DATE']))


#Create a Selectbox - for data display
visualization = st.sidebar.selectbox('select a chart type', ('Bar Chart','Pie Chart','Line Chart'))
supplier_select = st.sidebar.selectbox('Select a Supplier', data['SUPPLIER '].unique())
sensor_select = st.sidebar.radio(('Sensors'),('QZS','QZF','IPSQ (Re-work mins)','PPM - Scrap Rate',	'SOH',	'Supplier Internal Scrap Rate',	'MTTR',	'MTBF',	'Supplier Agreed Material stock',	'Supplier Agreed Finished Goods stock',	'Countermeasures',	'Due Date',	'Progress',	'Prognosis'))
selected_supplier = data[data['SUPPLIER ']==supplier_select]
st.markdown("## **Supplier level Analysis**")

cols1,_ = st.columns((1,2)) # To make it narrower
format = 'DD MM, YYYY'  # format output
start_date = dt.date(year=2023,month=11,day=1)-relativedelta(years=2)  #  I need some range in the past
end_date = dt.datetime.now().date()-relativedelta(years=2)
max_days = end_date-start_date

slider = cols1.slider('Select a date range', min_value=start_date, value=end_date ,max_value=end_date, format=format)
## Sanity check
st.table(pd.DataFrame([[start_date, slider, end_date]],
                      columns=['start',
                               'selected',
                               'end'],
                      index=['DATE']))


#Create a Selectbox - for data display
visualization = st.sidebar.selectbox('select a chart type', ('Bar Chart','Pie Chart','Line Chart'))
supplier_select = st.sidebar.selectbox('Select a Supplier', data['SUPPLIER '].unique())
sensor_select = st.sidebar.radio(('Sensors'),('QZS','QZF','IPSQ (Re-work mins)','PPM - Scrap Rate',	'SOH',	'Supplier Internal Scrap Rate',	'MTTR',	'MTBF',	'Supplier Agreed Material stock',	'Supplier Agreed Finished Goods stock',	'Countermeasures',	'Due Date',	'Progress',	'Prognosis'))
selected_supplier = data[data['SUPPLIER ']==supplier_select]
st.markdown("## **Supplier level Analysis**")
```
